```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from torch.utils.data import DataLoader, TensorDataset
```

```python
# 讀取 CSV 檔案
file_path = "2330_20100101_20241231_technical.csv"
df = pd.read_csv(file_path)

# 計算每日收益率（returns）
df["returns"] = df["收盤價"].pct_change()

# 計算 30 天移動標準差作為波動性（volatility）
df["volatility"] = df["returns"].rolling(window=30).std()

# 顯示計算結果
print(df[["日期", "收盤價", "returns", "volatility"]].head(50))  # 先看前 50 筆數據
# 假設我們要使用數值型的列來做 Autoencoder


df = df.dropna().reset_index(drop=True)
df.to_csv(file_path, index=False)


print(df.head())
```

```python
import pandas as pd

# 定義檔案名稱和對應的股票代碼
files = {
    "2317": "2317_20100101_20241231_technical.csv",
    "2330": "2330_20100101_20241231_technical.csv",
    "2454": "2454_20100101_20241231_technical.csv"
}

# 建立一個空的 DataFrame 來存放合併後的資料
df_list = []

for stock_code, file in files.items():
    df = pd.read_csv(file)  # 讀取 CSV 檔案
    df["股票代碼"] = stock_code  # 新增股票代碼欄位
    df_list.append(df)  # 加入合併列表

# 合併所有 DataFrame
merged_df = pd.concat(df_list, ignore_index=True)

# 根據日期排序（確保日期是時間格式）
merged_df["日期"] = pd.to_datetime(merged_df["日期"])  # 轉換為 datetime 格式
merged_df = merged_df.sort_values(by=["日期", "股票代碼"]).reset_index(drop=True)  # 按日期 & 股票代碼排序

# 儲存合併後的 CSV
merged_df.to_csv("merged_stocks_sorted（新增降維數據）.csv", index=False)

# 顯示前幾列資料
print(merged_df.head())
```
